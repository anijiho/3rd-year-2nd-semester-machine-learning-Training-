# 데이터 준비하기

```python
import pandas as pd
fish = pd.read_csv('https://bit.ly/fish_csv_data') # 쉼표로 구분된 값 csv
fish.head() # 상위 5개만 참조하겠다.
```

```python
fish_input = fish[['Weight','Length','Diagonal','Height','Width']].to_numpy()
print(fish_input[:5])
print(pd.unique(fish['Species']))
fish_target = fish['Species'].to_numpy() # 넘파이 배열로 바꾸는 이유는 ?? 
```

```python
from sklearn.model_selection import train_test_split
train_input,test_input,train_target,test_target = train_test_split(fish_input,fish_target,random_state = 42) # 전처리를 해줌 전치러를 하기 위해서는 스탠다드
```

```python
# 훈련세트 테스트 세트 분할
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(train_input)
train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)
```

```python
from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier(n_neighbors=3)
kn.fit(train_scaled,train_target)
print(kn.score(train_scaled,train_target))
print(kn.score(test_scaled,test_target))

print(kn.classes_) # 0번부터 6번 클래스까지로 인식
print(pd.unique(fish['Species'])) # 7개의 클래스로 처리
```

# 최근접 이웃 분류기의 결과 분석 

```python
print(kn.predict(test_scaled[:5]))
```

```python
import numpy as np
proba = kn.predict_proba(test_scaled[:5])
print(np.round(proba,decimals=4)) #round 는 반올림 
```

```python
print(kn.classes_)
```

# 시그모이드 함수 그리기

```python
import numpy as np
import matplotlib.pyplot as plt
z = np.arange(-5,5,0.1) # 어레인지 a 보다 크거나 같고 b보다 작거나 같다.
phi = 1 / (1+np.exp(-z)) # 상수 e 를 사용라는 수식 np.exp(상수 e의 계수)
plt.plot(z,phi)
plt.xlabel('z')
plt.ylabel('phi')
plt.show()
# 다중 분류 문제
```

# 이진 분류 체험하기

```python
bream_smelt_indexes = (train_target == 'Bream') | (train_target == 'Smelt')
train_bream_smelt = train_scaled[bream_smelt_indexes]
target_bream_smelt = train_target[bream_smelt_indexes]
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(train_bream_smelt, target_bream_smelt)
```

```python
print(lr.predict(train_bream_smelt[:5]))
print(lr.predict_proba(train_bream_smelt[:5]))
print(lr.classes_)
```

```python
print(lr.coef_,lr.intercept_)
```

```python
decisions = lr.decision_function(train_bream_smelt[:5])
print(decisions)
```

```python
from scipy.special import expit
print(expit(decisions))
```

```python
print(lr.predict_proba(train_bream_smelt[:5]))
```

# 다중 분류

```python
lr = LogisticRegression(C=20,max_iter=1000)
lr.fit(train_scaled,train_target)
print(lr.score(train_scaled,train_target))
print(lr.score(test_scaled,test_target))
```

```python
print(lr.predict(test_scaled[:5]))
```

```python
print(lr.classes_)
```

```python
proba = lr.predict_proba(test_scaled[:5])
print(np.round(proba,decimals=3))
```

```python
print(lr.coef_.shape,lr.intercept_.shape)
```

```python
decision = lr.decision_function(test_scaled[:5])
print(np.round(decision,decimals=2))
```

```python
from scipy.special import softmax
proba = softmax(decision,axis=1)
print(np.round(proba,decimals=3))
```
