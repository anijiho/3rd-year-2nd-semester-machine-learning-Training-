```python
import pandas as pd
wine = pd.read_csv('https://bit.ly//wine_csv_data')
```

```python
data = wine[['alcohol','sugar','pH']].to_numpy()
target = wine['class'].to_numpy()
```

```python
from sklearn.model_selection import train_test_split
train_input,test_input,train_target,test_target = train_test_split(data,target,test_size=0.2,random_state = 42)
```

```python
sub_input,val_input,sub_target,val_target = train_test_split(train_input,train_target,test_size=0.2,random_state=42)
```

```python
print(sub_input.shape,val_input.shape)
```

# 모델 훈련과 검증

```python
from sklearn.tree import DecisionTreeClassifier #결정트리분류 
dt = DecisionTreeClassifier(random_state = 42)
dt.fit(sub_input,sub_target)  # 서브 트리를 훈련
print(dt.score(sub_input,sub_target)) # 훈련세트
print(dt.score(val_input,val_target)) # 검증세트
```

# 교차 검증 구현 

```python
from sklearn.model_selection import cross_validate
scores = cross_validate(dt,train_input,train_target)
print(scores)
```

```python
import numpy as np
print(np.mean(scores['test_score']))
```

# 분할기

```python
from sklearn.model_selection import cross_validate
scores = cross_validate(dt,train_input,train_target)
```

```python
from sklearn.model_selection import StratifiedKFold
scores = cross_validate(dt,train_input,train_target,cv = StratifiedKFold())
```

```python
splitter = StratifiedKFold(n_splits=10,shuffle=True,random_state=42)
scores = cross_validate(dt,train_input,train_target,cv = splitter)
print(np.mean(scores['test_score']))
```

# 그리드 서치

```python
from sklearn.model_selection import GridSearchCV
params = {'min_impurity_decrease':[0.0001,0.0002,0.0003,0.0004,0.0005]}
```

```python
gs = GridSearchCV(DecisionTreeClassifier(random_state=42),params,n_jobs=-1)
```

```python
gs.fit(train_input,train_target)
```

```python
dt = gs.best_estimator_
print(dt.score(train_input,train_target))
```

```python
print(gs.best_params_)
```

```python
print(gs.cv_results_['mean_test_score'])
```

```python
best_index = np.argmax(gs.cv_results_['mean_test_score'])
print(gs.cv_results_['params'][best_index]) # 0.0001이 가장 좋은 모델이라는 것을 출력해주는 것
```
